```python
def percConvert(tab):
  return (tab/float(tab[-1])*100).round(decimals=0)

print('Tabela Cruzada por total de COLUNAS (em %):')
#Use axis igual a 0
ct3=pd.crosstab(cabreuva.HORASNAINTERNETPARAJOGAR,cabreuva.HORASNAINTERNETPARAESTUDAR,margins=True,rownames=['HORASNAINTERNETPARAJOGAR'], colnames=['HORASNAINTERNETPARAESTUDAR']).apply(percConvert, axis=0)
ct3
```
